```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# Replace 'your_data.csv' with your file path or source
data = pd.read_csv('C:/Users/Dell/Desktop/nypd-motor-vehicle-collisions.csv')  # Load data (if in CSV format)
data.head()  # Display the first few rows of the data




```

```python
# Data verification, description, and preparation
print(data.info())
print(data.isnull().sum())
```

```python
data.describe()
```

```python
# 1) Determination of the most dangerous factors of accidents in each NY district
dangerous_factors = data.groupby(['BOROUGH', 'CONTRIBUTING FACTOR VEHICLE 1']).size().reset_index(name='COUNT')
dangerous_factors = dangerous_factors[dangerous_factors['CONTRIBUTING FACTOR VEHICLE 1']!='Unspecified']
most_dangerous_factors = dangerous_factors.sort_values(['BOROUGH', 'COUNT'], ascending=[True, False])
most_dangerous_factors = most_dangerous_factors.groupby('BOROUGH').head(1)
print(most_dangerous_factors)
```

```python
# 1) Determination of the most dangerous factors of accidents in each NY district
dangerous_factors = data.groupby(['BOROUGH', 'CONTRIBUTING FACTOR VEHICLE 2']).size().reset_index(name='COUNT')
dangerous_factors = dangerous_factors[dangerous_factors['CONTRIBUTING FACTOR VEHICLE 2']!='Unspecified']
most_dangerous_factors = dangerous_factors.sort_values(['BOROUGH', 'COUNT'], ascending=[True, False])
most_dangerous_factors = most_dangerous_factors.groupby('BOROUGH').head(1)
print(most_dangerous_factors)
```

```python
# 2) Number of deaths and injuries caused by fast driving in a given district
fast_driving_stats = data.loc[data['CONTRIBUTING FACTOR VEHICLE 1'] == 'Unsafe Speed',['BOROUGH','NUMBER OF PERSONS KILLED', 'NUMBER OF PERSONS INJURED']]\
    .groupby('BOROUGH').sum()[['NUMBER OF PERSONS KILLED', 'NUMBER OF PERSONS INJURED']]
print(fast_driving_stats)
```

```python
# 3) 3 most common factors of accidents by district and city-wide
common_factors_district = dangerous_factors.groupby('BOROUGH')\
    .apply(lambda x: x.nlargest(3, 'COUNT')).reset_index(drop=True)

common_factors_city = data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(3)
print(common_factors_district)
print(common_factors_city)
```

```python
# 4) Vehicles most often involved in accidents
vehicle_types = data['VEHICLE TYPE CODE 1'].value_counts().head(10)
print(vehicle_types)
```

```python
# 5) Statistics of the number of accidents for each district
accidents_by_district = data['BOROUGH'].value_counts()

# Convert to DataFrame for better readability
accidents_by_district_df = accidents_by_district.reset_index()
accidents_by_district_df.columns = ['BOROUGH', 'NUMBER_OF_ACCIDENTS']

# Display the statistics
print("Statistics of the Number of Accidents by District:")
print(accidents_by_district_df)

# Visualization of the number of accidents by district
plt.figure(figsize=(10, 6))
accidents_by_district.plot(kind='bar', color='skyblue')
plt.title('Number of Accidents by District')
plt.ylabel('Number of Accidents')
plt.xlabel('District')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()








```
